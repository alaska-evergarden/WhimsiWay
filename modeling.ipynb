```python
import pandas;
import numpy;
import seaborn as sns;
import matplotlib.pyplot as plt
import plotly.express as px
import folium
from folium.plugins import HeatMap
```

```python
df = pandas.read_csv('tx_accidents.csv')
df.head()
```

```python
df.dropna()
df.drop_duplicates()
df = df.drop(["Unnamed: 0", "ID", "Source", "End_Lat", "End_Lng", "Civil_Twilight", "Nautical_Twilight", "Astronomical_Twilight"], axis=1)
df.head()
```

```python
color_scale = [(0, 'orange'), (1,'red')]

fig = px.scatter_mapbox(df, 
                        lat="Start_Lat", 
                        lon="Start_Lng",
                        hover_data="Description",
                        zoom=8, 
                        height=800,
                        width=800)

fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
# Create a folium map centered around the mean of the coordinates
map_center = [df['Start_Lat'].mean(), df['Start_Lng'].mean()]
mymap = folium.Map(location=map_center, zoom_start=5)

# Add a heatmap layer
HeatMap(df[['Start_Lat', 'Start_Lng']], radius=10, blur=5).add_to(mymap)

# Save the map as an HTML file
mymap.save("heatmap.html")
```
